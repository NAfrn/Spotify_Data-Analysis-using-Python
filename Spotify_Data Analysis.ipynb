```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tracks.csv')
df.head()
```

```python
df.info()
```

```python
#checking null values
pd.isnull(df).sum()
```

```python
#describtive statistics
df.describe().transpose()
```

```python
#finding 10 least popular songs in the dataset
sorted_df=df.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
#finding top 10 Most popular songs
sorted_popular_df= df.sort_values('popularity', ascending=False).head(10)
sorted_popular_df
```

```python
#finding top 10 Most popular songs with popularity greater than 90

most_popular=df.query('popularity>90', inplace = False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
# set index to release date and apply changes to data set by inplace = True

df.set_index('release_date',inplace=True)
df.head()
```

```python
#set the date format for index(release_date)
df.index=pd.to_datetime(df.index)
df.head()
```

```python
#filter to specific rows of info from dataset
#check the artist on 18th row use index location method iloc
df[['artists']].iloc[18]
```

```python
#convert duration from milliseconds to seconds
df['duration']=df['duration_ms'].apply(lambda x: round(x/1000))
df.head()
```

```python
#drop duration_ms and apply changes to dataset by inplace = True
df.drop('duration_ms', inplace=True, axis=1)
```

```python
df.head()
```

```python
df.duration.head()
```

```python
#Visualization
#drop unwanted columns - 'key', 'mode', 'explicit' 
#apply pearson corelation method
```

```python
corr_df = df.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize = (14,6))
#sns.heatmap(corr_df, annot=True)
heatmap = sns.heatmap(corr_df, annot = True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor = 'Black')
#set tittle for heatmap
heatmap.set_title('corelation Heatmap Between variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#create 0.4% sample data from the original data and then create 2 regression plots with this data
```

```python
sample_df= df.sample(int(0.004*len(df)))
print(len(sample_df))
```

```python
#create regression plot between loudness and energy
#In corelation map we saw there was positive corelation between loudness and energy
#lets plot it in the form of regression line
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= 'loudness', x= 'energy', color='c',line_kws={'color': 'red'}).set(title='Loudness vs Energy Corelation')
```

```python
#plot for popularity and acousticness
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= 'popularity', x= 'acousticness', color='b',line_kws={'color': 'red'}).set(title='Popularity vs Acousticness')
```

```python
#from above plot we can say if acousticness increases popularity decreases
#we can see the downword trend in the regression line
```

```python
# Create new column year from release date column
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
years
```

```python
# create distribution plot to visualize total songs in each year since 1992
```

```python
#update seaborn library
#!pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of songs per year')
```

```python
#above plot tells us that the number of songs for each year in the dataset have increased since music became more accesable through technological advancements
```

```python
# duration of songs over the years - bar plot
```

```python
total_dr= df.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title='years vs Duration')
plt.xticks(rotation=90)
```

```python
#from above barchart we can see that earlier in 1920's  the duration of the songs were less
#and later it increased around 1930's and this remain consistant untill 2010.
#after 2010 we can see the duration of the songs decresing could be becouse the attention span of listeners is decreasing
```

```python
#next create line plot to analyse average duration of songs over the years
```

```python
total_dr=df.duration
sns.set_style(style='whitegrid')
fig_dims=(10, 5)
fig, ax= plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr, ax=ax).set(title='Year vs Duration')
plt.xticks(rotation=60)
```
